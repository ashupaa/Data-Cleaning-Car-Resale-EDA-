```python
df['full_name']=df['full_name'].str.strip()
```

```python
df['Car_Model_Year']=df['full_name'].str.extract('(?i)(\d+)')
```

```python
df
```

```python
df[['Car_Company','Car_Model',"Car_Variant"]]=df['full_name'].str.split(pat=' ',expand=True).loc[:,1:3]
```

```python
df
```

```python
df.applymap(lambda x : str(x).strip())
```

```python
df['registered_year']=df['registered_year'].astype('string')
```

```python
df['Registered_Month']=df['registered_year'].str.extract(r'(^\w+)').replace('[0-9]','',regex=True)
```

```python
df['registered_year'].dtypes
```

```python
df
```

```python
df.rename(columns={'Registered_Month':'Car_Registered_Month', 'city':'City'})
```

```python
df.columns
```

```python
df['Car_Resale_Price(Lakh)']=df['resale_price'].str.strip('₹Lakh')
```

```python
df
```

```python
df
```

```python
df['Car_Engine_Capacity(CC)']=df['engine_capacity'].str.extract('(\d+)')
```

```python
df['Car_Kilometer_Driven(kms)']=df['kms_driven'].str.extract('(\d+)')
```

```python
df['Car_Max_Power(bhp)']=df['max_power'].str.extract('(\d+)')
```

```python
df['Car_Mileage(kmpl)']=df['mileage'].str.extract('(\d+)')
```

```python
df.drop(columns=['registered_year','engine_capacity','kms_driven','max_power','mileage','full_name','resale_price'],inplace=True)
```

```python
df.columns
```

```python
df = df[['Car_Company', 'Car_Model',
       'Car_Variant', 'Registered_Month', 'Car_Resale_Price(Lakh)',
       'Car_Engine_Capacity(CC)', 'Car_Kilometer_Driven(kms)',
       'Car_Max_Power(bhp)', 'Car_Mileage(kmpl)', 'Car_Model_Year','insurance', 'transmission_type', 'owner_type', 'fuel_type', 'seats',
       'body_type', 'city' ]]
```

```python
df.columns
```

```python
cols_to_clean = ['Car_Resale_Price(Lakh)', 'Car_Engine_Capacity(CC)',
                 'Car_Kilometer_Driven(kms)', 'Car_Max_Power(bhp)', 'Car_Mileage(kmpl)']

def to_float_or_none(val):
    try:
        return float(val) if not pd.isna(val) else None
    except:
        return None

df[cols_to_clean] = df[cols_to_clean].applymap(to_float_or_none)
```
